```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
scores = {'Algos':['RFC', 'XGB','LGBM','CatB','VC'], 'Specificity':[], 'Sensitivity':[], 'Precision':[], 'F1':[]}
```

####Combined Dataset

```python
scores['Specificity'] = [97.82, 97.43, 97.59, 97.35, 97.53]
scores['Sensitivity'] = [84.40, 85.84, 87.04, 85.92, 86.35]
scores['Precision'] = [89.97, 88.50, 88.61, 88.18, 88.69]
scores['F1'] = [86.81, 86.91, 87.45, 86.80, 87.26]
```

```python
results = pd.DataFrame(scores)
resm = results.melt('Algos', var_name='Metrics', value_name='values')
resm
```

```python
sns.catplot(x='Algos', y="values", hue="Metrics", data=resm, kind="point").set(title="Combined Data")
```
